***function that performs text normalization steps and returns the lemmatized tokens as a sentence***

```python
def text_normalization(text):
    text = str(text).lower()
    spl_char_text = re.sub(r'[^ a-z]','',text)      # removing special characters
    tokens = nltk.word_tokenize(spl_char_text)      # word tokenizing
    lema = wordnet.WordNetLemmatizer()
    tags_list = pos_tag(tokens,tagset=None)         # parts of speech
    lema_words = []
    for token,pos_token in tags_list:               # lemmatize according to POS
        if pos_token.startswith('V'):               # Verb
            pos_val = 'v'
        elif pos_token.startswith('J'):             # Adjective
            pos_val = 'a'
        elif pos_token.startswith('R'):             # Adverb
            pos_val = 'r'
        else:
            pos_val = 'n'                           # Noun
        lema_token = lema.lemmatize(token,pos_val)

        if lema_token in stop:
          lema_words.append(lema_token)             # appending the lemmatized token into a list

    return " ".join(lema_words)
```

```python
%%writefile app.py
import streamlit as st
import pandas as pd
import numpy as np
import re
import nltk
from nltk.stem import wordnet
from sklearn.feature_extraction.text import CountVectorizer, TfidfVectorizer
from nltk import pos_tag
from sklearn.metrics import pairwise_distances
from nltk import word_tokenize
from nltk.corpus import stopwords
import requests

# Download necessary NLTK resources
nltk.download('punkt')
nltk.download('wordnet')
nltk.download('averaged_perceptron_tagger')
nltk.download('stopwords')

# Function for text normalization
def text_normalization(text):
    text = str(text).lower()
    spl_char_text = re.sub(r'[^ a-z]','',text)
    tokens = nltk.word_tokenize(spl_char_text)
    lema = wordnet.WordNetLemmatizer()
    tags_list = pos_tag(tokens, tagset=None)
    lema_words = []
    for token, pos_token in tags_list:
        if pos_token.startswith('V'):
            pos_val = 'v'
        elif pos_token.startswith('J'):
            pos_val = 'a'
        elif pos_token.startswith('R'):
            pos_val = 'r'
        else:
            pos_val = 'n'
        lema_token = lema.lemmatize(token, pos_val)
        if lema_token not in stop:
            lema_words.append(lema_token)
    return " ".join(lema_words)

# Load data
path_to_csv = '/content/mentalhealth.csv'
df = pd.read_csv(path_to_csv, nrows=20)

# Preprocessing
stop = stopwords.words('english')
df['lemmatized_text'] = df['Questions'].apply(text_normalization)

# Initialize CountVectorizer
cv = CountVectorizer()
df_bow = cv.fit_transform(df['lemmatized_text']).toarray()

# Initialize TF-IDF Vectorizer
tfidf = TfidfVectorizer()
x_tfidf = tfidf.fit_transform(df['lemmatized_text']).toarray()

# Define functions for Bag of Words and TF-IDF chat
def chat_bow(text):
    lemma = text_normalization(text)
    bow = cv.transform([lemma]).toarray()
    cosine_value = 1 - pairwise_distances(df_bow, bow, metric='cosine')
    index_value = cosine_value.argmax()
    max_score = cosine_value.max()
    return df['Answers'].loc[index_value], "bow_bot", max_score

def chat_tfidf(text):
    lemma = text_normalization(text)
    question_tfidf = tfidf.transform([lemma]).toarray()
    cos = 1 - pairwise_distances(x_tfidf, question_tfidf, metric='cosine')
    index_value = cos.argmax()
    max_score = cos.max()
    return df['Answers'].loc[index_value], "tfidf_bot", max_score

# Hugging Face API URL
API_URL = "https://api-inference.huggingface.co/models/mistralai/Mixtral-8x7B-Instruct-v0.1"
headers = {"Authorization": "Bearer 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"}

# Query function for Hugging Face API
def query(payload):
    response = requests.post(API_URL, headers=headers, json=payload)
    if response.status_code == 200:
        return response.json()
    else:
        st.error(f"Error: {response.status_code}")
        st.error(f"Response: {response.text}")
        return None

# Define function for generative chat
def chat_generative(text):
    output = query({"inputs": text, "max_length": 150})  # Increase max_length for more text
    if output:
        return output[0]['generated_text'], "generative_bot"

# Streamlit UI
def chat_interface():
    st.title("Psychological Well-Being Chatbot")
    st.write("Welcome! \n"
         "Psychological well-being is a person's overall functioning and emotional health, which includes:\n"
         "- Positive social relationships\n"
         "- Autonomy\n"
         "- Environmental mastery\n"
         "- Self-acceptance\n"
         "- Sense of purpose\n"
         "- Personal growth\n"
         "/nI'm here to help you with your questions/concerns.")


    # Dropdown for selecting chat method
    chat_method = st.selectbox("Select Chat Method:", ("Bag of Words (BoW)", "TF-IDF", "Generative"))
    user_input = st.text_input("You:")
    if st.button("Ask"):
        if chat_method == "Bag of Words (BoW)":
            response, method_name, max_score = chat_bow(user_input)
        elif chat_method == "TF-IDF":
            response, method_name, max_score = chat_tfidf(user_input)
        else:
            response, method_name = chat_generative(user_input)
            max_score = None

        if max_score:
            st.write(f"{method_name}:", response)
            st.write(f"{method_name} Max Score:", max_score)
        else:
            st.write(f"{method_name}:", response)

    # Display helplines below the chat interface
    st.markdown("""
    ---
    ### Emergency Helplines
    - 988 Suicide & Crisis Lifeline -> Call 988
    - NAMI HELPLINE -> Text 'HelpLine' to 62640 or call 800-950-NAMI (6264)
    - Substance Abuse and Mental Health Services Administration (SAMHSA) National Helpline -> Call 1-800-662-4357
    """)

if __name__ == "__main__":
    chat_interface()
```
